```python
from ultralytics import YOLO
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
model = YOLO("/content/drive/MyDrive/DatasetSnowPole/Runs_iphone/v1-16/weights/last.pt")
model.train(
    data="/content/drive/MyDrive/DatasetSnowPole/Roadpole_betterformat/data.yaml",
    epochs=250,
    imgsz=1920,
    mosaic=1.0,
    batch=24,
    project="/content/drive/MyDrive/DatasetSnowPole/Runs_iphone",
    name="v1_continued",
    resume=True
)
```

```python
model = YOLO("/content/drive/MyDrive/DatasetSnowPole/Runs_iphone/v1-16/weights/best.pt")
model.predict(
    source="/content/drive/MyDrive/DatasetSnowPole/Roadpole_betterformat/test/images",
    project="/content/drive/MyDrive/DatasetSnowPole/Predictions_iphone",
    name="submission_v1_19",
    augment=True,
    conf=0.0001,
    iou=0.05,
    save_txt=True,
    save_conf=True # <--- This adds the probability of each predicted box
    )
```

```python
import shutil

shutil.make_archive(
    "/content/drive/MyDrive/DatasetSnowPole/Predictions_iphone/submission_v1_19",  # output zip path
    "zip",
    "/content/drive/MyDrive/DatasetSnowPole/Predictions_iphone/submission_v1_19/labels"  # folder to zip
)
```
