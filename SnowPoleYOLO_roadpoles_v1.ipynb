## Training model

```python
model = YOLO("yolo11s.pt")
model.train(
    data="/content/drive/MyDrive/DatasetSnowPole/roadpoles_v1/data.yaml",
    epochs=400,
    imgsz=1920,
    batch=12,
    mosaic=1.0,
    project="/content/drive/MyDrive/DatasetSnowPole/Runs",
    name="v21"
)
```

```python
model = YOLO("/content/drive/MyDrive/DatasetSnowPole/Runs/v13/weights/best.pt")
model.predict(
    source="/content/drive/MyDrive/DatasetSnowPole/roadpoles_v1/test/images",
    project="/content/drive/MyDrive/DatasetSnowPole/Predictions",
    name="submission_v21_5",
    save_txt=True,
    conf=0.01,
    iou = 0.3,
    save_conf=True # <--- This adds the probability of each predicted box
    )
```

```python
import shutil

shutil.make_archive(
    "/content/drive/MyDrive/DatasetSnowPole/Predictions/submission_v21_5",  # output zip path
    "zip",
    "/content/drive/MyDrive/DatasetSnowPole/Predictions/submission_v21_5/labels"  # folder to zip
)
```
